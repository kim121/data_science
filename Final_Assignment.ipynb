# 그래프들에 대한 비교, 분석

### 기본적으로 classic그래프들과 random그래프들의 차이점에서 가장 두드려저 보이는 것은 clustering coefficient이다. 이하 clustering coefficient = CC ,  그래프에 대한 Average clustering coefficient=ACC로 지칭
### 클래식 그래프들은 대칭적이거나 규칙적인 그래프들이기 때문에 CC가 모든 노드들이 비슷하거나 패턴이 있고 그렇기 때문에 ACC또한 0이나 1이 자주 보인다. 
### 그러나 random 그래프들의 경우에는 어떠한 규칙은 있다하더라도 랜덤적인 요소가 강하기 때문에 (노드의 연결에 확률적인 부분이 있으므로) CC가 노드들마다 굉장히 불규칙적으로 다르다. ACC 또한 마찬가지로 굉장히 무작위적인 숫자가 나오고 그래프를 새로 그릴때마다 이 숫자들은 달라진다.  

### radius나 diameter, center, periphery의 경우엔 classic이나 random이 아니라 각 그래프들의 개별적인 특성에 따라 다르기 때문에 어떠한 공통점이 존재하지 않는다고 생각한다. 







# 2. Twitter 소셜 그래프 특성 분석

### http://snap.stanford.edu/data/egonets-Twitter.html
### 위 사이트의 데이터셋을 가지고 트위터 소셜 그래프를 분석해보기.
### 분석을 위해 snap 라이브러리를 추가하여 snap 메서드를 사용

```python
import csv
import snap
from scipy import sparse, stats
```

```python
twt=snap.LoadEdgeList(snap.PNGraph, "twitter_combined.txt",0,1)
```

```python
snap.PrintInfo(twt, "QA Stats", "twt-info.txt", False)
```

```python
# %load twt-info.txt
QA Stats: Directed
  Nodes:                    81306
  Edges:                    1768149
  Zero Deg Nodes:           0
  Zero InDeg Nodes:         1
  Zero OutDeg Nodes:        11209
  NonZero In-Out Deg Nodes: 70096
  Unique directed edges:    1768149
  Unique undirected edges:  1342310
  Self Edges:               14
  BiDir Edges:              851692
  Closed triangles:         13082506
  Open triangles:           190845756
  Frac. of closed triads:   0.064152
  Connected component size: 1.000000
  Strong conn. comp. size:  0.841426
  Approx. full diameter:    7
  90% effective diameter:  4.564534
```

```python
MxWcc = snap.GetMxWcc(twt)
print "max wcc nodes %d, edges %d" %(MxWcc.GetNodes(), MxWcc.GetEdges())
```

```python
MxScc = snap.GetMxScc(twt)
print "max scc nodes %d, edges %d" %(MxScc.GetNodes(), MxScc.GetEdges())
```

```python
NId = snap.GetMxDegNId(twt)
print "max degree node", NId
```

```python
CC=snap.GetClustCf(twt)
print "clustering coefficient", CC
```

```python
ED = snap.GetBfsEffDiam(twt, 100)
print "effective diameter", ED
```

#### 총 노드의 수 : 81306
#### 총 간선의 수 : 1768149
#### Average Clustering coefficient :  대략 0.565311
#### Diameter : 7
#### Effective diameter : 약 4.58742
#### 최대 WCC안의 총 노드의 수 : 81306
#### 최대 WCC안의 총 간선의 수 : 1768149
#### 최대 SCC안의 총 노드의 수 : 68413
#### 최대 SCC안의 총 간선의 수 : 1685163

#### 과제 내용 중 radius는 관련 메서드를 찾을 수 없어서 구할 수 없었음
#### 또한 centrality는 snap내에서 PUN그래프에만 적용가능하고 PN그래프에는 적용 불가능해 구할 수 없었음.

#### 결과를 분석해보면 이 트위터 네트워크에는 총 81306명의 사용자가 존재하며 그들 사이에 간선은 1768149개이다. ACC가 0.565311 이다라는 것은 어떤 유저 A가 있을 때 그 A유저의 친구들끼리도 연결되있을 확률이 꽤 있다고 볼 수 있다. Diameter가 7이기 때문에 모든 유저는  7단계만 거치면 연결되있고 , Effective Diameter(90%의 유저로 봤을때, 즉 특별히 혼자 동떨어진 아웃라이어를 제거했을 경우)가 약 4.6이기때문에 실제론 대략 4.6단계만 거치면 연결되있다는 것이다. 
#### 최대 WCC의 노드와 간선의 수는 총 노드,간선의 수와 같기 때문에 즉 이 트위터 네트워크는 하나의 WCC라고 볼 수있다
#### 최대 SCC를 살펴보았을때 노드의 수가 68413, 간선의 수가 1685163으로, 네트워크 전체에서 꽤 큰 비율로 존재한다. 즉 이 트위터 네트워크는 거의 강 결합이라고 보아도 좋을 듯 하다.
